```python
# Importing the dataset Click Through Rate optimization
dataset = pd.read_csv('Ads_Optimization.csv')
print(dataset.shape)
dataset.head()
```

```python
reward_sum=[]
for i in range(0,10):
    reward_sum.append(dataset.iloc[:,i].sum())
    
plt.bar(x=list(range(10)), height = reward_sum)
plt.title('Bar Chart')
plt.xlabel('Ads')
plt.ylabel('Reward Sum')
plt.show()
```

```python
# Implementing Random Selection
import random
random.seed(10)
N = 10000
d = 10
```

```python
ads_selected = []
total_reward = 0
for n in range(0, N):
    ad = random.randrange(d)
    ads_selected.append(ad)
    reward = dataset.values[n, ad]
    total_reward = total_reward + reward
```

```python
total_reward # with random selection
```

```python
# Visualising the results
plt.hist(ads_selected, color='orange')
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```

```python
# Implementing UCB
import math
N = 10000
d = 10
ads_selected = []
numbers_of_selections = [0] * d
sums_of_rewards = [0] * d
total_reward = 0
for n in range(0, N):
    ad = 0
    max_upper_bound = 0
    for i in range(0, d):
        if (numbers_of_selections[i] > 0):
            average_reward = sums_of_rewards[i] / numbers_of_selections[i]
            delta_i = math.sqrt(3/2 * math.log(n + 1) / numbers_of_selections[i])
            upper_bound = average_reward + delta_i
        else:
            upper_bound = 1e400
        if upper_bound > max_upper_bound:
            max_upper_bound = upper_bound
            ad = i
    ads_selected.append(ad)
    numbers_of_selections[ad] = numbers_of_selections[ad] + 1
    reward = dataset.values[n, ad]
    sums_of_rewards[ad] = sums_of_rewards[ad] + reward
    total_reward = total_reward + reward
```

```python
# Visualising the results
plt.hist(ads_selected, color='violet')
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
